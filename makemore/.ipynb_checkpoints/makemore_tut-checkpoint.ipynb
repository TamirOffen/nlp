```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
```

## dataset

```python
# load names dataset 
words = open('names.txt', 'r').read().splitlines()
print(f'number of examples(names) is {len(words)}')
print(f'example names: {words[:3]}')
print(f'length of shortest name: {min(len(w) for w in words)}, \
length of longest name: {max(len(w) for w in words)}')
```

## bigram

```python
b = {} # dict(bigram) = count
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
```

```python
sorted(b.items(), key = lambda kv: -kv[1])[:5]
```

we want to work with a tensor instead of dict

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0 # special token
itos = {i:s for s,i in stoi.items()}
```

```python
N = torch.zeros((26+1,26+1), dtype=torch.int32)
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        idx1, idx2 = stoi[ch1], stoi[ch2]
        N[idx1, idx2] += 1
```

counts array N visualization 

```python
plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues')
for i in range(27):
    for j in range(27):
        chstr = itos[i] + itos[j]
        plt.text(j, i, chstr, ha='center', va='bottom', color='gray') #bigram possibilities 
        plt.text(j, i, N[i, j].item(), ha='center', va='top', color='black') #counts
_ = plt.axis('off')
```

```python
# making the first row into a prob. dist.
p = N[0].float()
p = p/p.sum()
p
```

```python
g = torch.Generator().manual_seed(2147483647)
ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
itos[ix]
```

full name generation using the bigram count matrix N

```python
# normalized counts N
P = (N+1).float() #+1 so we dont have counts of 0 (model smoothing, the more we add, the "smoother" the probs)
P /= P.sum(dim=1, keepdim=True)
# according to broadcasting rules: 27,27 / 27,1 = 27, 27
# 27, 27    27, 27
# 27,  1 => 27, 27 the column gets copied 27 times
P[0].sum()
```

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(10): #num of generated names
    out = []
    ix = 0 #start at special token .
    while True: #gen name
        # p = N[ix].float()
        # p = p/p.sum()
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0: # end token (special token .)
            break
    print(''.join(out))
```

### Bigram Model Evaluation

```python
log_likelihood = 0.0 
n = 0
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        idx1, idx2 = stoi[ch1], stoi[ch2]
        prob = P[idx1, idx2]
        logprob = torch.log(prob)
        log_likelihood += logprob; n += 1
#         print(f'{ch1}{ch2}: {prob:.4f} -log-> {logprob:.4f}')

print(f'{log_likelihood=}')
nll = -log_likelihood
print(f'{nll=}')
print(f'avg nll (loss) = {nll/n}') # the lower the better
```

## Neural Net Model
a very simple nn of one linear layer that takes in 27 inputs and has 27 neurons, with no activation function, and a softmax.

input char -stoi-> char index -model-> model output which is a tensor of length 27 which is a prob. dist. over the next char.

```python
# training set for the nn model, which is the bigrams (x, y)
xs, ys = [], []

for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        idx1, idx2 = stoi[ch1], stoi[ch2]
        xs.append(idx1)
        ys.append(idx2)
        
xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()
print(f'number of samples: {num}')
```

network init

```python
W = torch.randn((27,27), requires_grad=True) #each column is a neuron, each neuron receives 27 inputs.
# (batch_size,27) @ (27,27) -> (batch_size,27)
```

gradient descent

```python
for k in range(120):
    # forward pass
    xenc = F.one_hot(xs, num_classes=27).float() # input to network, (B, 27)
    logits = xenc @ W # nn output, log counts, (B, 27) #note1
    counts = logits.exp() # equivalent to N, (B, 27)
    probs = counts / counts.sum(dim=1, keepdims=True) # probs of next char (P), (B, 27)
    loss = -probs[torch.arange(num), ys].log().mean() + (W**2).mean() # nll + reg (note2)
    print(loss.item())
    
    # backward pass
    W.grad = None # zero out the gradients
    loss.backward()
    
    # update
    W.data += -50 * W.grad
    
```

```python
#note1: ohe vector just gets the corresponding row of W, just like in the bigram approach.
#note2: reg term acts like the smoothing in the bigram approach, selecting for a more unifrom prob. dist. 
#       When W is 0, we get a uniform dist. for each row in P.
```
